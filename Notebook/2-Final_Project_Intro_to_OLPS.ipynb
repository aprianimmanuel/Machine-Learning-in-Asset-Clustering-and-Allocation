```python
import pandas as pd
```

```python
#import csv files and value sorting based on Equity

agri = pd.read_csv('Agriculture.csv')
chem = pd.read_csv('Basic Industry and Chemicals.csv')
consumer = pd.read_csv('Consumer.csv')
fin = pd.read_csv('Finance.csv')
infra = pd.read_csv('Infrastructure.csv')
mining = pd.read_csv('Mining.csv')
misc = pd.read_csv('Miscellaneous Industry.csv')
prop = pd.read_csv('Property.csv')
trade = pd.read_csv('Trade.csv')

l = [agri, chem, consumer, fin, infra, mining, misc, prop, trade]
df = pd.concat(l,keys= ['agri', 'chem', 'consumer', 'fin', 'infra', 'mining', 'misc', 'prop', 'trade'],axis=0).reset_index()
df_drop = df.drop(['level_0', 'level_1', 'Unnamed: 0'], axis=1)
df_drop.sort_values(by='Equity', ascending=False)
```
